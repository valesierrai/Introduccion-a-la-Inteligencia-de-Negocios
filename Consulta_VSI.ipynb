#2. Se ejecuta SQLite

```python
#1. Se crea/lee la base de datos
conexion = sqlite3.connect('ConsultaSectorEnergia.db')
cursor = conexion.cursor()
```

```python
#Dado que se tienen varias tablas, usaremos un ciclo para pasar la información desde el diccionario hasta la base de datos creada
for hoja, datos in datos.items():
  datos.to_sql(hoja, conexion, if_exists = 'replace', index = False)
```

```python
#Confirmamos los cambios ejecutados en la base de datos
conexion.commit()
```

#3. Creamos la Consulta

```python
#Creamos la consultas (formato de texto), en SQL
#Para visualizar los nombres de todas las tablas que tiene la base de datos
consulta = '''SELECT name
           FROM sqlite_master
           WHERE type = 'table' '''

#Ejecutamos la consulta
cursor.execute(consulta)

#Mostramos los resultados
print(cursor.fetchall()) #fetchall: Se trae toda la información de la base de datos para visualizar
```

Queremos saber con base al ID y nombre de la empresa, de aquellas empresas que se encuentren en Antioquia y NO tengan Subsidios

```python
#Primero seleccionamos las variables
#Segundo, cuando aplicamos el FROM, le agregamos EM y CT1 después de los nombres de las tablas, para poder unir información de dos tablas en una sola consulta y que el programa me lo reconozca
#Sabiendo, que me va a unir ambas tablas por medio del ID
#Finalmente, hacemos los filtros que queremos, dejandole saber al programa que ANTIOQUIA es un texto
consulta = '''
SELECT EM.ID || ' ' || EM.Empresa AS InfoEmpresa,
       EM.Departamento AS Sede,
       CT2."Total Subsidios" AS Subsidios
       FROM 'Info empresas' EM
       JOIN 'Características 2' CT2 ON EM.ID = CT2.ID
       WHERE Sede = 'ANTIOQUIA' AND Subsidios = 0 '''

#Ejecutamos la consulta
cursor.execute(consulta)

#Mostramos los resultados
resultados = cursor.fetchall()
for resultado in resultados:
  print(resultado)
```
